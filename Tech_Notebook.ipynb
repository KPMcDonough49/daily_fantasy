```python
#BASE_URL="http://rotoguru1.com/cgi-bin/fyday.pl?game=fd&scsv=1&week=WEEK&year=YEAR"#fanduel
BASE_URL="http://rotoguru1.com/cgi-bin/fyday.pl?game=fd&scsv=1&week=WEEK&year=YEAR"#Draft kings
WEEKS=list(map(str,range(1,18)))
YEARS=list(map(str,range(2017,2020)))
```

```python
all_games=pd.DataFrame()
for yr in YEARS:
    for wk in WEEKS:
        soup=utils.soup(BASE_URL.replace("WEEK",wk).replace("YEAR",yr))
        all_games=pd.concat([all_games,pd.read_csv(io.StringIO(soup.find("pre").text),sep=";")])
all_games
```

```python
all_games.to_csv('Data/web_scrape.csv', index=False)
```
